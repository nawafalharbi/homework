# Get data from the provided URL

```python
url='http://archive.ics.uci.edu/ml/machine-learning-databases/postoperative-patient-data/post-operative.data'
#download the file 
raw_data= urlopen(url)
```

# Load the CSV File 

```python
dataset = np.genfromtxt(raw_data, delimiter=",",dtype=str)
```

```python
dataset.shape
```

```python
# assign x= data , y=features 
```

```python
x=dataset[:,0:89]
y=dataset[0,8]
```

```python
x
```

# we have three missing vraiables indicated by 
(?) we need to regenrate it using median 

# display data in a table 

```python
x1=np.copy(x)
#we will replace the ? by nan
x1[x1=='?']=np.nan
x1
```

```python
import pandas as pd
dataframe=pd.DataFrame(x1,columns=[ 'L-CORE','L-SURF','L-O2','L-BP','SURF-STBL','CORE-STBL','BP-STBL','COMFORT','decision ADM-DECS'])
dataframe
```

# it is easy to process the data using numbers instead of strings. 
so we will transfer all strings to numbers



```python
size_mapping={'low':1 , 'mid':5 , 'high':10}
dataframe['L-CORE']=dataframe['L-CORE'].map(size_mapping)
dataframe['L-SURF']=dataframe['L-SURF'].map(size_mapping)
dataframe['L-BP']=dataframe['L-BP'].map(size_mapping)

dataframe
```

```python
size_mapping={'poor':6 , 'fare':7 ,'good':8, 'excellent':9}
dataframe['L-O2']=dataframe['L-O2'].map(size_mapping)
dataframe
```

```python
size_mapping={'stable':3 , 'mod-stable':2 ,'unstable':1, }
dataframe['SURF-STBL']=dataframe['SURF-STBL'].map(size_mapping)
dataframe['CORE-STBL']=dataframe['CORE-STBL'].map(size_mapping)
dataframe['BP-STBL']=dataframe['BP-STBL'].map(size_mapping)
dataframe
```

```python
size_mapping={'I':33 , 'S':22 , 'A':11}
dataframe['decision ADM-DECS']=dataframe['decision ADM-DECS'].map(size_mapping)
dataframe
```

# now we replace each 'nan' (missing variables by a regenerated variable using median) mx=missing variables 

```python
from sklearn.preprocessing import Imputer 
mx=Imputer(missing_values='NaN',strategy='median',axis=0)
mx=mx.fit(dataframe)
my=mx.transform(dataframe.values)
my
```

SHOW DATA AGAIN IN TABLE FORM

```python
import pandas as pd
dataframe=pd.DataFrame(my,columns=[ 'L-CORE','L-SURF','L-O2','L-BP','SURF-STBL','CORE-STBL','BP-STBL','COMFORT','decision ADM-DECS'])
dataframe
```

# after we see the statistical window. we would like to see if there is a relationship between variables. 

# we use  the pearson correlation r and p-value  between each features comapred to the comfort .

#this method gives us a quick view of how each variable related to the comfort. 

```python
import pandas 
import numpy 
import seaborn 
import scipy
import matplotlib.pyplot as plt
scat1=seaborn.regplot(x='L-CORE',y='COMFORT',fit_reg=True, data=dataframe )
plt.xlabel('L-CORE')
plt.ylabel('COMFORT')
plt.title('scatterplot for L-CORE and COMFORT')
print ('correlation and p-value for L-CORE related to COMFORT')
print(scipy.stats.pearsonr(dataframe['L-CORE'],dataframe['COMFORT']))
```

```python
#as we can see that from the correlation approximately 0.01 
#this number means there is no linear relationship between L-CORE and COMFORT.
#and p-value is greater than 0.05 which means it is not statistically significant. 
```
